```python
import pandas as pd

data_frames = {
    "gpt4turbo_ru": pd.read_excel("drive/MyDrive/chat_api/get_option/gpt4turbo_ru.xlsx"),
    "gpt4turbo_en": pd.read_excel("drive/MyDrive/chat_api/get_option/gpt4turbo_en.xlsx"),
    "gpt4o_ru": pd.read_excel("drive/MyDrive/chat_api/get_option/gpt4o_ru.xlsx"),
    "gpt4o_en": pd.read_excel("drive/MyDrive/chat_api/get_option/gpt4o_en.xlsx"),
    "gpt35_ru": pd.read_excel("drive/MyDrive/chat_api/get_option/gpt35_ru.xlsx"),
    "gpt35_en": pd.read_excel("drive/MyDrive/chat_api/get_option/gpt35_en.xlsx"),
    "gemini_en": pd.read_excel("drive/MyDrive/chat_api/get_option/gemini_en.xlsx"),
    "gemini_ru": pd.read_excel("drive/MyDrive/chat_api/get_option/gemini_ru.xlsx"),
    "mistral_en": pd.read_excel("drive/MyDrive/chat_api/get_option/mistral_en.xlsx"),
    "mistral_ru": pd.read_excel("drive/MyDrive/chat_api/get_option/mistral_ru.xlsx"),
    "llama_en": pd.read_excel("drive/MyDrive/chat_api/get_option/llama_en.xlsx"),
    "llama_ru": pd.read_excel("drive/MyDrive/chat_api/get_option/llama_ru.xlsx")
}

# Создание списка для хранения данных
results = []

# Обход всех датасетов
for model_name, df in data_frames.items():
    # Добавляем колонку, фиксирующую правильность ответа
    df["is_correct"] = (df["extracted_letter"] == df["Answer"]).astype(int)
    print(len(df))
    df["model"] = model_name
    # Группировка по секциям с расчётом метрик
    section_stats = df.groupby("model").agg(
        mean_BERTScore_F1=pd.NamedAgg(column="BERTScore_F1", aggfunc="mean"),
        mean_ROUGE1_F1=pd.NamedAgg(column="ROUGE1_F1", aggfunc="mean"),
        mean_ROUGE2_F1=pd.NamedAgg(column="ROUGE2_F1", aggfunc="mean"),
        accuracy_corr=pd.NamedAgg(column="is_correct", aggfunc=lambda x: x.mean() * 100),  # В процентах
        accuracy_not_corr=pd.NamedAgg(column="is_correct", aggfunc=lambda x: 100-x.mean() * 100)
    ).reset_index()

    # Добавление колонки с названием модели
    section_stats["model"] = model_name

    # Добавление данных в общий список
    results.append(section_stats)

# Объединение всех данных в одну таблицу
comparison_df = pd.concat(results, ignore_index=True)

# Вывод таблицы
from IPython.display import display
display(comparison_df)
```

```python
import pandas as pd

# Загружаем данные
df = comparison_df  # Подставь свою переменную с таблицей

# Создаём сводную таблицу
pivot_table = df.pivot_table(
    index="section",
    columns="model",
    values=["mean_BERTScore_F1", "mean_ROUGE1_F1", "mean_ROUGE2_F1", "accuracy"],
    aggfunc="mean"
)

# Убираем вложенные заголовки для удобства
pivot_table.columns = [f"{metric}_{model}" for metric, model in pivot_table.columns]

# Экспортируем таблицу в LaTeX без booktabs
latex_code = pivot_table.to_latex(float_format="%.3f", column_format="|l|" + "c|" * len(pivot_table.columns), escape=False)

# Сохраняем в файл
with open("pivot_table.tex", "w") as f:
    f.write(latex_code)

# Вывод кода таблицы для проверки
print(latex_code)
```
